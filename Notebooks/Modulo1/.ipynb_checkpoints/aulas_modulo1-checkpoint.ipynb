### Desafio 6: Adicionar uma coluna mostrando a região de cada estado

Para adicionar a região verifiquei o código que acompanhava o nome de cada um dos estados e com isso adicionei a região correspondente.

```python
regiao = []
for i in dados_atendimento.index:
    i = int(i[:2])
    if i < 21:
        regiao.append('norte')
    if i > 17 and i < 31:
        regiao.append('nordeste')
    if i > 30 and i < 41:
        regiao.append('sudeste')
    if i > 40 and i < 44:
        regiao.append('sul')
    if i > 44 and i < 54:
        regiao.append('centro_oeste')
dados_atendimento['regiao'] = regiao
dados_atendimento['regiao']
```

```python
dados_atendimento['Total'] = dados_atendimento.sum(axis=1)
dados_atendimento.Total
```

```python
ordenado_total = dados_atendimento.sort_values(by='Total', ascending=False)
ordenado_total.drop('Total', axis=1, inplace=True)
ordenado_total.head()
```

### Desafio 7: Deixe o gráfico mais bonito e a mensagem mais clara

```python
ax = ordenado_total[:5].T.drop('regiao').plot(figsize=(10,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
ax.set_title('5 estados que mais gastaram com atendimento hospitalar até Nov/2021')
plt.grid(alpha=.5)
plt.show()
```

# Bloco 04: Análise de gráficos e criação de hipóteses

```python
ordenado_total_ = ordenado_total.drop('regiao', axis=1) / 1000000
ordenado_total_.iloc[:5,6:].T.plot(figsize=(10,6))
```

```python
meses_filtrados = ordenado_total_.columns[53:154]
ordenado_total_.head(3)[meses_filtrados].T.plot(figsize=(10,6)) 
plt.show()
```

```python
ordenado_total_.head(3)[meses_filtrados].T.plot(figsize=(10,6)) 
plt.ylim(0,350)
plt.show()
```

### Cuidado!
Quando cortamos um gráfico na horizontal ou vertical estamos limitando nossa visão e com isso deturpamos alguma informação. Uma espécie de ilusão é criado enviesando a conclusão a ser tomada.

```python
#exemplo de gráfico deturpado onde os gastos de São Paulo são 6x mais do que os gastos do Paraná
ordenado_total_.head(3)[meses_filtrados].T.plot(figsize=(10,6)) 
plt.ylim(100,350)
plt.show()
```

```python
# exemplo de gráfico deturpado que passa a idea de os crescimentos estão acontecendo de forma rápida no decorrer do tempo
ax = ordenado_total_.head(5)[meses_filtrados].T.plot(figsize=(6,6)) 
ax.legend(bbox_to_anchor=(1,1))
```
